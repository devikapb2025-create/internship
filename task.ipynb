```python
import pandas as pd
import numpy as np
```

```python
data = [
    [2001, 'C101', '2023-08-01 18:00', '2023-08-01 18:45', 500, 'upi', 'Dominos', 4],   ###create the dataset
    [2002, 'C102', '2023-08-01 19:00', '2023-08-01 18:50', 700, 'Credit card', 'KFC', 5],
    [2003, 'C103', '2023/08/02 20:00', np.nan, np.nan, 'Debit Card', 'McDonalds', 3],
    [2004, 'C104', 'invalid', '2023-08-03 21:00', 800, 'UPI', 'Dominos', 6],
    [2005, 'C105', '2023-08-04 18:30', '2023-08-04 19:00', -200, 'Cash', 'KFC', 2],
    [2005, 'C105', '2023-08-04 18:30', '2023-08-04 19:00', -200, 'Cash', 'KFC', 2],
    [2006, 'C106', '2023-08-05 17:00', '2023-08-05 17:40', 1200, '', 'Pizza Hut', 4],
]

columns = ['order_id', 'customer_id', 'order_time', 'delivery_time', 'order_value', 'payment_method', 'restaurant', 'rating']
df = pd.DataFrame(data, columns=columns) ##load
df
```

```python
df['order_time'] = pd.to_datetime(df['order_time'], errors='coerce')
df['delivery_time'] = pd.to_datetime(df['delivery_time'], errors='coerce')
df                                                                 ## clean data
```

```python
df = df.dropna(subset=['order_time', 'delivery_time']) ## remove invalid timestamps
df
```

```python
df = df[df['delivery_time'] >= df['order_time']]## delivery time >= order time
df
```

```python
df = df.drop_duplicates() ##remove duplicate orders
df
```

```python
df = df[df['order_value'] > 0] ##handle negative or missing order values
df
```

```python
df = df[(df['rating'] >= 1) & (df['rating'] <= 5)] ##fix ratings
df
```

```python
df['payment_method'] = df['payment_method'].str.lower().str.strip()

df['payment_method'] = df['payment_method'].replace({                  ###standardize payment methods
    'upi': 'UPI',
    'credit card': 'Credit Card',
    'debit card': 'Debit Card',
    'cash': 'Cash',
    '': 'Unknown'
})
df
```

```python
df.reset_index(drop=True, inplace=True)
df
```
